```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('naivebayes.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded = le.fit_transform(df['Outlook'])
df['Outlook'] = encoded
encoded = le.fit_transform(df['Temperature'])
df['Temperature'] = encoded
encoded = le.fit_transform(df['Humidity'])
df['Humidity'] = encoded
encoded = le.fit_transform(df['Windy'])
df['Windy'] = encoded
encoded = le.fit_transform(df['Play Golf'])
df['Play Golf'] = encoded
df
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(columns = 'Play Golf', axis = 1)
y = df['Play Golf']
print("Shape of X: ", X.shape)
print("Shape of y: ", y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 10)
print("Shape of X_train: ", X_train.shape)
print("Shape of X_test: ", X_test.shape)
print("Shape of y_train: ", y_train.shape)
print("Shape of y_test: ", y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
gnb_pred = gnb.predict(X_test)
print("Accuracy Score of Naive Bayes: ", accuracy_score(y_test, gnb_pred))
print("Roll Num: CB.SC.I5DAS21049")
print("Name: Pratiksha Cauvery K P")
```
