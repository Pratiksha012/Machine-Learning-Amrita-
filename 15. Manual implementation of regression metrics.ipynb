```python
import pandas as pd
df = pd.read_csv('Student_Performance_regression.csv')
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
encoded = le.fit_transform(df['Extracurricular Activities'])
df['Extracurricular Activities'] = encoded
df
```

```python
from sklearn.model_selection import train_test_split
import numpy as np

X = df.drop(columns = 'Performance Index')
Y = df['Performance Index']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)

from sklearn.neighbors import KNeighborsRegressor
knn_regressor = KNeighborsRegressor(n_neighbors=5)
knn_regressor.fit(X_train, y_train)

y_pred = knn_regressor.predict(X_test)

mae = np.mean(np.abs(y_test - y_pred))
print("Mean Absolute Error:", mae)

mse = np.mean((y_test - y_pred) ** 2)
print("Mean Squared Error:", mse)

mape = np.mean(np.abs((y_test - y_pred) / y_test)) * 100
print("Mean Absolute Percentage Error:", mape)

rmse = np.sqrt(np.mean((y_test - y_pred) ** 2))
print("Root Mean Square Error:", rmse)

r2 = 1 - (np.sum((y_test - y_pred) ** 2) / np.sum((y_test - np.mean(y_test)) ** 2))
print("R2 Score:", r2)
print("Roll Num: CB.SC.I5DAS21049")
print("Name: Pratiksha Cauvery K P")
```
