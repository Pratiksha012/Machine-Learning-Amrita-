```python
import pandas as pd
from sklearn import preprocessing, utils
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
X = df.drop('target', axis=1)
Y = df['target']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size=0.8)
print("Shape of X: ",X.shape)
print("Shape of Y",Y.shape)
print("Shape of X_train",X_train.shape)
print("Shape of X_test",X_test.shape)
print("Shape of Y_train",Y_train.shape)
print("Shape of Y_test",Y_test.shape)
```

# Gaussian Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

gnb = GaussianNB()
gnb.fit(X_train, Y_train)
gnb_pred  = gnb.predict(X_test)
print('Predicted Value:', gnb_pred)
print('Actual Value:', Y_test)
print("Accuracy Score: ", accuracy_score(Y_test,gnb_pred))
print("Roll Num: CB.SC.I5DAS21049")
print("Name: Pratiksha Cauvery K P")
```

# Support Vector Machines

```python
from sklearn.svm import SVC

svm = SVC(kernel = 'linear')
svm.fit(X_train, Y_train)
svm_pred = svm.predict(X_test)
print('Predicted Value:', svm_pred)
print('Actual Value:', Y_test)
print("Accuracy Score: ", accuracy_score(Y_test,svm_pred))
print("Roll Num: CB.SC.I5DAS21049")
print("Name: Pratiksha Cauvery K P")
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()
dtc.fit(X_train,Y_train)
dtc_pred = dtc.predict(X_test)
print('Predicted Value:', dtc_pred)
print('Actual Value:', Y_test)
print("Accuracy Score: ", accuracy_score(Y_test,dtc_pred))
print("Roll Num: CB.SC.I5DAS21049")
print("Name: Pratiksha Cauvery K P")
```

# K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
knn.fit(X_train, Y_train)
knn_pred = knn.predict(X_test)
print('Predicted Value:', knn_pred)
print('Actual Value:', Y_test)
print("Accuracy Score: ", accuracy_score(Y_test, knn_pred))
print("Roll Num: CB.SC.I5DAS21049")
print("Name: Pratiksha Cauvery K P")
```
