```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
df = pd.read_csv('Linear.csv')
df
```

```python
def estimate_coef(x, y):
    n = np.size(x)
    m_x = np.mean(x)
    m_y = np.mean(y)
    SS_xy = n*np.sum(y*x) -(np.sum(x)*np.sum(y))
    SS_xx = n*np.sum(x*x) - (np.sum(x)*np.sum(x)) 
    m = SS_xy / SS_xx
    c = (np.sum(y) - m*np.sum(x))/n
    return (m,c)
def plot_regression_line(x, y, m, c):
    plt.scatter(x, y, color = "m", marker = "o", s = 30)
    y_pred = (m*x)+c
    plt.plot(x, y_pred, color = "g") 
    plt.xlabel('Height')
    plt.ylabel('Weight')
    plt.title('Simple Linear Regression')
    plt.show()
if __name__ == "__main__":
    x = df['Height']
    y = df['Weight']
    m, c = estimate_coef(x, y)
    print("Estimated coefficients: Slope =", m, " Intercept =", c)
    plot_regression_line(x, y, m, c)
print("Roll Num: CB.SC.I5DAS21049")
print("Name: Pratiksha Cauvery K P")
```
